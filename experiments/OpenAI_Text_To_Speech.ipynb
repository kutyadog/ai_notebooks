```python
from google.colab import userdata
import ipywidgets as widgets
from IPython.display import display
```

# Text to Speech Settings

**Important:** Make sure you add your api key from https://platform.openai.com/api-keys to secrets. (Click the key icon on the left and copy paste your key in OPENAI_API_KEY).

Please note that OpenAI's usage policies require you to provide a clear disclosure to end users that the TTS voice they are hearing is AI-generated and not a human voice.

## Choosing a model

TTS is cheaper but TTS-HD is better. In some cases, the audio may not have noticeable differences depending on your listening device and the individual person.

At time of writing (25 Dec 2023):

**TTS: 0.015 USD per 1k chars**

**TTS-HD: 0.030 per 1k chars**

##Audio Samples, Pricing, Docs##

Always refer OpenAI website for any pricing changes and model updates.

Text to speech docs: https://platform.openai.com/docs/guides/text-to-speech

Pricing page: https://openai.com/pricing

```python
# Text input field
text_input = widgets.Textarea(
    placeholder='Type something here...',
    rows=6,  # Adjust the number of rows as needed
    layout=widgets.Layout(width='80%')  # Adjust the width as needed
)

# Dropdown for voice selection
voice_dropdown = widgets.Dropdown(
    description='Voice:',
    options=['alloy',
            #  'echo', 'fable', 'onyx',
             'nova', 'shimmer'],
    value='alloy'  # default value
)

# Dropdown for model selection
model_dropdown = widgets.Dropdown(
    description='Model:',
    options={'TTS': 'tts-1', 'TTS HD': 'tts-1-hd'},
    value='tts-1-hd'
)

# Label to display expected cost
cost_label = widgets.Label()
display(text_input, voice_dropdown, model_dropdown, cost_label)
```

```python
text_length = len(text_input.value)
cost_per_1k_chars = 0.015 if model_dropdown.value == 'tts-1' else 0.030
estimated_cost = (text_length / 1000) * cost_per_1k_chars
cost_label.value = f"Estimated Cost: ${estimated_cost:.4f}"
```

# Generate audio
Click play to generate speech from text you entered.

Then click Files icon on the left and download speech.mp3. Or just right click on the player and "save audio as".


```python
import openai
from IPython.display import Audio

# Set up OpenAI API key
client = openai.OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

# Generate audio
response = client.audio.speech.create(
    model=model_dropdown.value,
    voice=voice_dropdown.value,
    input=text_input.value
)

# Save the audio to a file and play it
file_name = 'speech.mp3'
response.stream_to_file(file_name)
display(Audio(file_name, autoplay=False))
```
